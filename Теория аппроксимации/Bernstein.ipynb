```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.special
import matplotlib.pyplot as plt
import math

%matplotlib inline

sns.set(style="darkgrid")
```

```python
def f(x):
    return math.exp(x)
```

```python
def C(n, k):
    return scipy.special.binom(n, k)
```

# Многочлен Бернштейна на [0, 1]

```python
def BernsteinPolynomial (x, n = 3):
    sm = 0
    for k in range(0, n + 1):
        sm += f(k/n) * C(n, k) * (x**k) * (1-x)**(n-k)
    return sm
```

```python
cnt = 100
n = 15
x = np.linspace(0, 1, num = cnt)
y = np.zeros(cnt)
```

```python
for i in range(0, cnt):
    y[i] = BernsteinPolynomial(x[i], n)
```

```python
plt.figure(figsize = (20,8), dpi = 80)
plt.plot(x, y, label='Многочлен Бернштейна')
plt.plot(x, np.exp(x), label='Апроксимируемый многочлен')
plt.legend(loc='upper left', frameon=False)
```

# Многочлен Бернштейна на [a, b]

```python
def BernsteinPolynomialProd(x, a, b, n = 3):
    sm = 0
    for k in range(0, n + 1):
        sm += f(a + ((b - a)*k)/n) * C(n, k) * ((x - a)**k) * (b-x)**(n-k)
    return sm / (b - a)**n
```

```python
cnt = 100
n = 15
a = 1
b = 3
x = np.linspace(a, b, num = cnt)
y = np.zeros(cnt)
```

```python
for i in range(0, cnt):
    y[i] = BernsteinPolynomialProd(x[i], a, b, n)
```

```python
plt.figure(figsize = (20,8), dpi = 80)
plt.plot(x, y, label='Многочлен Бернштейна')
plt.plot(x, np.exp(x), label='Апроксимируемый многочлен')
plt.legend(loc='upper left', frameon=False)
```
