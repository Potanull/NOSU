### Решение уравнения Лапласа
Мы ищем решение $u(x,y)$ уравнения Лапласа
$$\Delta u := \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0, \quad 0<x,y<1$$
с граничными условиями Дирихле
$$u(x,y) = f(x,y), \qquad \mbox{for } x=0, x=1, y=0, \mbox{ or } y=1$$
для заданной функции $f(x,y)$, определенной на граница квадрата $[0,1]\times[0,1]$. Сначала мы загружаем необходимые модули Python.

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LightSource
```

Далее мы определяем три функции. Функция `jacobiIterations` решает дискретизированное уравнение Лапласа, используя схему итераций Якоби, которая останавливается, когда последующие обновления становятся достаточно малыми. Функция `boundaries` корректирует обновления нашей итерационной схемы, чтобы убедиться, что они удовлетворяют заданным граничным условиям. Наконец, функция `graphSolution` визуализирует решение в виде поверхностного графика и контурного графика.

```python
# Определите граничные условия
def boundaries(p, f):
    p[:, 0]  = f[0] # левая сторона x=0
    p[:, -1] = f[1] # правая сторона x=1
    p[0, :]  = f[2] # нижняя сторона y=0
    p[-1, :] = f[3] # верхняя сторона y=1
    return p

# Решите уравнение Лапласа
def jacobiIterations(p, dx, dy, f, l1norm_target):
    l1norm = 1
    pn = np.empty_like(p)
    p = boundaries(p, f)
    while l1norm > l1norm_target:
        pn = p.copy()
        p[1:-1, 1:-1] = ((dy**2 * (pn[1:-1, 2:] + pn[1:-1, 0:-2]) +
                          dx**2 * (pn[2:, 1:-1] + pn[0:-2, 1:-1])) /
                          (2 * (dx**2 + dy**2)))
        p = boundaries(p, f)
        l1norm = (np.sum(np.abs(p[:]) - np.abs(pn[:])) /
                np.sum(np.abs(pn[:])))
    return p

# Сюжетное решение
def graphSolution(u, x, y):
    X,Y = np.meshgrid(x, y)
    light = LightSource(180, 45)
    illuminated_surface = light.shade(u, cmap=cm.coolwarm)
    fig = plt.figure(figsize=(5, 10))
    ax0 = fig.add_subplot(211, projection='3d')
    ax0.plot_surface(X, Y, u, rstride=1, cstride=1, linewidth=0,
                     antialiased=False, facecolors=illuminated_surface)
    ax0.set_xlabel('x')
    ax0.set_ylabel('y')
    ax0.set_zlabel('u')
    ax0.set_title('Решение u(x,y)')
    ax0.view_init(30, 225)
    ax1 = fig.add_subplot(212)
    graph1 = ax1.imshow(u, cmap='coolwarm',  aspect='equal', origin='lower',
                        interpolation='bilinear', extent=[min(x), max(x), min(y), max(y)])
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.set_title('Решение u(x,y)')
    ax1.set_xlim(min(x), max(x))
    ax1.set_ylim(min(y), max(y))
    fig.colorbar(graph1, ax=ax1)
    plt.tight_layout()
    plt.show()
```

Теперь мы можем задать наши числовые параметры $nx$ и $ny$, которые определяют количество точек пространственной сетки, которые мы используем для дискретизации нашей области $[0,1]\times[0,1]$ в направлениях $x$ и $y$ соответственно. Затем мы определяем граничные условия, решаем уравнение Лапласа и визуализируем решение.

```python
# Объявлять числовые параметры
nx = 51
ny = 51

# Определите интервалы
dx = 1 / (nx - 1)
dy = 1 / (ny - 1)
x = np.linspace(0, 1, nx)
y = np.linspace(0, 1, ny)

# Определяет функцию f(x,y) на границе в порядке слева, справа, снизу и сверху
f = [np.sin(2*np.pi*y), np.sin(np.pi*y), np.sin(np.pi*x), np.sin(np.pi*x)]

# Решите уравнение Лапласа
p = np.zeros((ny, nx))
p = jacobiIterations(p, dx, dy, f, 1e-4)
graphSolution(p, x, y)
```
