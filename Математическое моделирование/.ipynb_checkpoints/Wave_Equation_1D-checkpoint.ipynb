### Волновое уравнение на ограниченных интервалах
Наша цель - вычислить и визуализировать решения волнового уравнения $u_{tt}=c^2 u_{xx}$ на интервале $0<x<L$ с начальными условиями $u(x,0)=f(x)$ и $u_t(x,0)=g(x)$. Наш код позволит нам использовать следующие граничные условия:
* Граничные условия Дирихле: $u(0,t)=0$ и $u(L,t)=0$;
* Граничные условия Неймана (или без потока): $u_x(0,t)=0$ и $u_x(L,t)=0$;
* Периодические граничные условия: $u(0,t)=u(L, t)$ и $u_x(0,t)=u_x(L,t)$, что помещает волновое уравнение на окружность окружности $L$.

Сначала мы загружаем необходимые модули Python.

```python
%matplotlib widget
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import matplotlib.colors as colors
```

Далее мы определяем следующие функции:
* `discrete_laplacian` обеспечивает центрированную конечно-разностную аппроксимацию второй производной $u_{xx}$;
* `update_solution` использует метод чехарды (центрированная конечно-разностная аппроксимация производной по времени $u_{tt}$) для обновления решения до следующего временного шага;
* `update_graph` анимирует решение;
* `define_initial_condition` использует заданные начальные условия для вычисления первых двух итераций решения.

```python
def discrete_laplacian(u, bdy):
    # периодические граничные условия:
    L = -2*u
    L += np.roll(u, -1)
    L += np.roll(u, +1)
    if bdy == 'Dirichlet':
        L[0]  = -2*u[0]+u[1]
        L[-1] = -2*u[-1]+u[-2]
    elif bdy == 'Neumann':
        L[0]  = 2*(-u[0]+u[1])
        L[-1] = 2*(-u[-1]+u[-2])
    return L

def update_solution(f, g, v, bdy, cfl, dt, Nframes, Nskip):
    n = 0
    u0, um = define_initial_condition(f, g, bdy, dt)
    v[0, :] = u0
    while n<Nframes:
        n += 1
        for k in range(Nskip):
            up = 2*u0 - um + cfl**2*discrete_laplacian(u0, bdy)
            um = u0
            u0 = up
        v[n, :] = u0
        yield u0

def update_graph(u, x, line1):
    line1.set_data(x, u)

def define_initial_condition(f, g, bdy, dt):
    um = f
    u0 = um + 0.5*cfl**2*discrete_laplacian(um, bdy) + dt*g
    return u0, um
```

В следующей ячейке мы определяем константу скорости $c>0$, задаем начальные условия $u(x,0)=f(x)$ и $u_t(x,0)=g(x)$ и запускаем анимацию решения. Исходным условием, используемым здесь, является единый локализованный профиль. Попробуйте следующую последовательность симуляций:
* Используйте f и g, как определено ниже.
* * Замените g на -g (например, убрав `#` перед `g=-g`).
* Замените g на ноль (например, убрав `#` перед `g=0*g`).

```python
# параметр модели
c = 1 # постоянная скорость
L = 10 # длина домена

# числовые параметры
N = 1000 # размер сетки

# установите граничные условия
bdy = 'Periodic'
if bdy=='Periodic':
    x = np.linspace(0, L, N, endpoint=False)
    dx = L/N
else:
    x = np.linspace(0, L, N)
    dx = L/(N-1)

# установите начальные условия
f = 1/np.cosh((x-L/3)*5)
g = 5*c*np.sinh((x-L/3)*5)/np.cosh((x-L/3)*5)**2
# g = -g
# g = 0*g

# параметры анимации
Nsteps = 4000
Nskip = 10

# установите временной шаг для удовлетворения критерия стабильности скачкообразного изменения (cfl<1)
cfl = 0.3
dt = cfl * dx / c

# компьютерное и анимационное решение
Nframes = int(Nsteps/Nskip)
v = np.zeros((Nframes, N))

fig = plt.figure()
ax  = plt.axes(xlim=(0,L), ylim=(np.min(np.min(f)*1.1,0), np.max(np.max(f)*1.1,0)))
line1, = ax.plot([], [], linewidth=2, color="tab:blue")
ax.set_xlabel('x')
ax.set_ylabel('u')
ani = animation.FuncAnimation(fig, update_graph,
                              update_solution(f, g, v, bdy, cfl, dt, Nframes, Nskip),
                              fargs=(x, line1), interval=10, blit=True, repeat=False)
plt.tight_layout()
plt.show()
```

Наконец, мы визуализируем решение с помощью пространственно-временного контурного графика.

```python
fig = plt.figure(figsize=(5,4))
plt.imshow(v, cmap='Blues', extent = [0, L, 0, Nsteps*dt], aspect='auto', origin='lower')
plt.title('u(x,t)')
plt.xlabel('x')
plt.ylabel('t')
plt.colorbar()
plt.tight_layout()
plt.show()
```
